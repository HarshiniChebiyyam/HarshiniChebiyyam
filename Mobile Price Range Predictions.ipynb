Splitting data: 

```python
#from sklearn.model_selection import train_test_split
X_train=train_data.drop(['price_range'],axis=1)
#print(X_train)
Y_train=train_data['price_range']
#print(Y_train)
```

```python
x_test=test_data.drop(['id'],axis=1)
#print(x_test)
```

Using classification technique for finding accuracy of the model.



Different types of classifiers are logistic regression, Random forest,svm ,Navie Bayes,Decision trees etc 
These algorithms used for supervised learning .

At first we go with logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(C=100,random_state=0)
lr=lr.fit(X_train,Y_train)
y_pred=lr.predict(test_data)
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(LogisticRegression(),X_train,Y_train,cv=10)
print(scores)
print(scores.mean())
```

```python
data={'id':sample_submission['id'],'price_range':y_pred}
result=pd.DataFrame(data)
result
result.to_csv('/kaggle/working/output.csv',index=False)  #index=false means that the index value is not written in the csv output
output=pd.read_csv('/kaggle/working/output.csv')
```

Here we are using svm (support vector machine)classifier algorithm

```python

from sklearn.svm import SVC
svm=SVC(C=1.0,kernel='linear',random_state=0)  #(Support Vector Classification - SVC())
                                              # kernel types are of linear,rbf,poly,sigmond etc.
svm=svm.fit(X_train,Y_train)
y_pred=svm.predict(test_data)
print(svm.score(X_train,Y_train))  #Return the mean accuracy on the given test data and labels
```
